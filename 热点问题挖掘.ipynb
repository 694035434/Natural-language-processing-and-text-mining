```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import jieba as jb
import re
df=pd.read_excel("C题全部数据/附件3.xlsx")
df
```

```python
df.isnull().sum()
```

```python
#定义删除除字母,数字，汉字以外的所有符号的函数
def remove_punctuation(line):
    line = str(line)
    if line.strip()=='':
        return ''
    rule = re.compile(u"[^a-zA-Z0-9\u4E00-\u9FA5]")
    line = rule.sub('',line)
    return line
 
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  
 
#加载停用词
stopwords = stopwordslist("停用词.txt")
```

```python
df['new_details'] = df['留言详情'].apply(remove_punctuation)
df['cut_details'] = df['new_details'].apply(lambda x: " ".join([w for w in list(jb.cut(x)) if w not in stopwords]))
df.head()
```

```python
labels=[]
for i in range(len(df)):
    labels.append(df.留言主题[i])
```

```python
df.cut_details.values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
# 该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频
vectorizer = CountVectorizer(max_features=20000)
# 该类会统计每个词语的tf-idf权值
tf_idf_transformer = TfidfTransformer()
# 将文本转为词频矩阵并计算tf-idf
tfidf = tf_idf_transformer.fit_transform(vectorizer.fit_transform(df.cut_details.values))
# 获取词袋模型中的所有词语
weight = tfidf.toarray()
# 获取词袋模型中的所有词语
word = vectorizer.get_feature_names()
```

```python
from sklearn.cluster import KMeans
mykms=KMeans(n_clusters=100)
y=mykms.fit_predict(weight)
class_dict={}
class_list=[]
for i in range(0,100):
    label_i=[]
    for j in range(0,len(y)):
        if y[j]==i:
            label_i.append(labels[j])
    class_dict['label_'+str(i)]=str(len(label_i))
    class_list.append(label_i)
    print('label_'+str(i)+':'+str(label_i))
```
