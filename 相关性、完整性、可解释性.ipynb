```python
import pandas as pd
# import numpy
df=pd.read_excel('data/附件4.xlsx')
df.head(5)
```

```python
# 选出要进行评估的两项数据
df=df[['留言详情','答复意见']]
df
```

```python
# 对留言详情进行处理
import re
def f1(x):
    result=re.sub("\s+","",x)
    return result
df.留言详情=df.留言详情.apply(f1)
df
```

```python
# 查看是否有空值
df.isnull().sum()
```

```python
#定义删除除字母,数字，汉字以外的所有符号的函数
def remove_punctuation(line):
    line = str(line)
    if line.strip()=='':
        return ''
    rule = re.compile(u"[^a-zA-Z0-9\u4E00-\u9FA5]")
    line = rule.sub('',line)
    return line
 
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  

#加载停用词
stopwords = stopwordslist("data/停用词表.txt")
```

```python
#删除除字母,数字，汉字以外的所有符号
df['留言详情'] = df['留言详情'].apply(remove_punctuation)
df["答复意见"]=df['答复意见'].apply(remove_punctuation)
df.sample(5)
```

```python
#分词，并过滤停用词
import jieba as jb
df['留言详情'] = df['留言详情'].apply(lambda x: " ".join([w for w in list(jb.cut(x)) if w not in stopwords]))
df['答复意见'] = df['答复意见'].apply(lambda x: " ".join([w for w in list(jb.cut(x)) if w not in stopwords]))
df.head()
```

```python
# 通过列表实现编辑距离算法计算两个数据集的相似度
def edit_distance(str1, str2):
    matrix = [[i+j for j in range(len(str2) + 1)] for i in range(len(str1) + 1)]
    for i in range(1,len(str1)+1):
        for j in range(1,len(str2)+1):
            if str1[i-1] == str2[j-1]:
                d = 0
            else:
                d = 1
            matrix[i][j] = min(matrix[i-1][j]+1,matrix[i][j-1]+1,matrix[i-1][j-1]+d)
    return matrix[len(str1)][len(str2)]

x_list=[]

for i in range(len(df)):
    a=edit_distance(df.iloc[i][0],df.iloc[i][1])
    b=1-a/(max(len(df.iloc[i][0]),len(df.iloc[i][1])))
    x_list.append(b)  
    
sum(x_list)/len(x_list) 
```
